$ \hat{\beta} = \frac{ \sum_{i=1}^{n} (x_i-\bar{x})(y_i - \bar{y})} {\sum_{i=1}^{n} (x_i - \bar{x})^2} $

$ \hat{\beta} = \frac{ \sum_{i=1}^{n} (x_i-0)(y_i - 0)} {\sum_{i=1}^{n} (x_i - 0)^2} $

$ \hat{\beta} = [ Cov[X,Y] ] / Var[X]  $

$ \hat{\beta} = [ Cov[X,Y] ] $

Both converge to $ [ Cov[X,Y] ]$, or just the covariance of X and Y. In a specific condition where $X$ and $Y$ are both standard normal distributions, they are actually the exact same statistic. 


%Question 4 Part A

$ RSE = \sigma_{model}^2 = \sqrt{ \frac{RSS}{n-2} } = \sqrt{ \frac{ \sum_{i=1}^{n} (y-\hat{y})^2 }{n-2} }$

RSE is a variance estimate of the residual error, and is the simplest form of the goodness of fit measures. It provides a measure of how much of the observed variance $\hat{y}$ explains


$ r^2 = \frac{TSS-RSS}{TSS} = 1 - \frac{RSS}{TSS} = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2} {\sum_{i=1}^{n} (y_i - \bar{y})^2}$ 

$ r^2 $ provides as measure of what percentage of the variance in Y is explained by X, it is nice in that in provides a goodness of fit measure that is interpretable and has meaningful and interpretable units. It is constrained between 0 and 1 with 0 suggesting very bad fit of model and 1 suggesting very good fit of model


$ F = \frac{ \frac{(TSS-RSS)}{p}} { \frac{RSS}{n - p - 1}} $

The F-statistic gives a proportion of the  variance not explained by your model (TSS) is relative to the variance explained your model (RSS)
