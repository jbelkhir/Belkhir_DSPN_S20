Picture Naming RT as a function of Intra-Operative Awake Stimulation

Humans have the amazing ability to have develop and maintain deeply structured mental models of the world that we live in. Consider the relatively simple example of a dog chasing a boy. Whether this model is comes from direct visual input, retrieval of a memory, or procured from imagination – the semantic components of the scene (i.e. the dog, the boy, the action) are all processed in tandem, and across the different cognitive domains, represent the same deep structure. Most of us use speech as a primary form of communication, that allows us to convey our own internal mental models to those around us. When we think about sentence planning, this by necessity happens at a scope that includes multiple words. Sentences contain content that cannot be specified to a single word, and at the syntactic level, we have to compute a whole grammatical structure before we begin an utterance. While these mental models with deep structure are abstracted in parallel processes, speech is temporally constrained in that we can only produce one sound at a time, and as such it is produced as a serialized motor output. 

So how does the language network solve the computational problem of deep structure in the context of speech production? We know that the language system is broadly organized. And we also speech is a motor action, so is therefore coordinated by the motor cortex, such that there is a one-to-one correspondence between activity of the speech motor cortex and the muscular components of speech production. Therefore, every neural component subserving speech production, in one way or another, by necessity must relay information to the speech motor cortex as a final point of convergence in the brain. In other words, all roads lead to the speech motor cortex, and this is given the nature of the direct projections through the brainstem, this functional area can be thought of as the last time that brain ‘touches’ language.

We want to understand the broad language network support the translation of temporally and spatially extended content, that is simultaneously represented in the system, to a serially produced speech as a motor action in the context of speech and lexical access. We leverage data collected from Direct Electical Stimulation brain mapping in a picture naming task during awake neurosurgery. We collect response time measures in the operating room for trials carried out during the stimulation mapping procedure. The idea is that we can leverage this data to map subprocesses of lexical access on the cortex using direct electrical stimulation. Considering speech motor cortex as the epicenter of the speech production process, the idea is to relate each point of stimulation in terms of structural to the speech motor cortex, and see how performance (RT) associated with the stimulation of these structures, is predicited as a function of connectivity to the speech motor cortex.

DATA ORGANIZATION

1. Select Data relevant for analysis
    - Patient # (n=13)
    - Trial number(n=1701)
    - Accuracy 
        - Accuracy: For current trials (n=1701)
        - accuracy_above: For trial preceding current trial (n=1701)
    - Psycholinguistic Variable
        - Repetition: Number of times individual has seen same image in the mapping session
        - Familiarity: Psycholinguistic value generated for each image corrsponding to average familiariy of item in English language
    - Stimulation Parameters
        - StimDuration (ms): Duration of time stimulating rod is in contact with cortical surface
        - TimeofStim (ms) : Timing of electical stimulation onset as a function of picture presentation. Negative values refer to stimulation that began before the picture stimulus was presented, positive values indicate the amount of time after picture presentation that electical stimulation was initiated. 
2. Remove Errors
    - Remove errors, and trials preceded by error (accuacy=1 for correct trials)
    - Remove trials with RT > 3500ms, that should have been coded as errors
3. Caluculate Zscore for each RT 
    - Calculate Zscore as a function of each individuals means and sd of RT from individuals non-stimulated trials
4. Remove unnecessary variable
    - Accuracy + accuracy_above all = 1
        - no longer needed
    - Baseline_mean and baseline_sd calculated to make Zscore_RT variable
        - no longer needed
5. Subset stimulated and non-stimulated data
6. Limit analyses of stim data
    - Only include trials that Stim_Duration is bw 1000-5000ms
    - Only include trials where stimulation starts with 1500ms of pic presentation, and before response
6. Create a binarized value of RT
    - for classification analyses
    - faster, or slower, relative to baseline RT
7. Summarize data


```r
# function to install the packages 

import_packages <- function(packages) {
  for (package in packages) {
    if (!(package %in% row.names(installed.packages()))) { 
      install.packages(package, repos = "http://cran.us.r-project.org") 
    } 
    library(package, character.only = TRUE) 
  }
}

import_packages(c('tidyverse'))
import_packages(c('tree'))
import_packages(c('randomForest'))
import_packages(c('ISLR'))
import_packages(c('ggplot2'))
import_packages(c('plot3D'))
import_packages(c('ggfortify'))
#import_packages(c('MASS'))

setwd("~/Desktop/Grad_School/Classes/Spring_2019/Classes/Data_Science/Final/")
d<-read.csv("AwakePicNaming_Data.csv")
library(tidyverse)
library(tree)
library(randomForest)
library(ISLR)

# Retrieve Variables of Interest

d %>%
  select( Trial, Patient, Accuracy, accuracy_above,
          Repetition,  Familiarity, Latency_Preamble,  StimDuration, TimeofStim)-> d1

# remove error trials and trials preceded by error and extremely slow pic-naming_RTs that may not be coded as errors

accurate_d1 <- d1  %>% 
  filter(Accuracy == 1,
         accuracy_above==1,
         Latency_Preamble<3500)

# calculate average RT per patient for non-stimulated trails (i.e. 'baseline' RT)
#and add column in data set measuring Z-scoreing RT relative to that patients non-stim 'baseline' RT to calibrate
#for variable RT distribution bw patient in the operating room

for (c in unique(accurate_d1$Patient))
{Patient_d <- accurate_d1[accurate_d1$Patient==c & accurate_d1$StimDuration=="NaN",]
for (x in 1:length(accurate_d1$Trial))
{if (accurate_d1$Patient [x] == c)
{accurate_d1$Baseline_RT [x] <- mean(Patient_d$Latency_Preamble);
accurate_d1$Baseline_SD [x] <- sd(Patient_d$Latency_Preamble);
accurate_d1$Zscore_RT [x] <- (accurate_d1$Latency_Preamble [x] - mean(Patient_d$Latency_Preamble))/mean(Patient_d$Latency_Preamble) 
# Z-score RT for all trials (Stim and non-stim) relative to mean and distribution of non-stim baseline trials for 
# each patient
}}}

#Remove columns that are no longer needed
    # Pic code is an arbitrary value for each different picture stimulus
    # Accuracy and accuracy_above all equal 1
    # Baseline_RT + _SD were calculated to caluculate Zscore_RT
    
accurate_d1 <- subset(accurate_d1, select = -c( Accuracy, accuracy_above, Baseline_RT, Baseline_SD))

# seperate Stimulated and non-stimulated trials

Stim_Trials <- accurate_d1 %>% 
  filter(StimDuration != 'NaN')
NonStim_Trials <- accurate_d1 %>% 
  filter(StimDuration == 'NaN')

# Filter Stim_Trials to trials where stimulation started within 1500ms of PicPresentation and before response
# and the stimulating rod was in contact with cortex bw 1000-5000ms and remove trial more than two distributions
# from mean

nrow(Stim_Trials)
Cleaned_Stim_Trials <- Stim_Trials %>% 
  filter(StimDuration>1000, StimDuration<5000,
         TimeofStim<Latency_Preamble, TimeofStim>-1500, Zscore_RT<2)
    nrow(Cleaned_Stim_Trials)



# Create binary variable for faster/slower RT relative to non-stim 'baseline' for classification analyses

Cleaned_Stim_Trials$Faster_RT<-ifelse((Cleaned_Stim_Trials$Zscore_RT>0), 'No', 'Yes')
Cleaned_Stim_Trials$Faster_RT <- factor(Cleaned_Stim_Trials$Faster_RT)


# All subsequent analyses will be restricted to this dataset "Cleaned_Stim_Trials", with Zscore_RT being the 
# transformed variable of analyses

# Preliminary Tree Decision visualization

tree.Cleaned_Stim_Trials <- tree(Zscore_RT~.-Latency_Preamble -Patient -Faster_RT -Trial, Cleaned_Stim_Trials)
tree.Cleaned_Stim_Trials
plot(tree.Cleaned_Stim_Trials)
text(tree.Cleaned_Stim_Trials, pretty=0) 
```

```r
# Regression analysis indicates that StimDuration and TimeofStim are primary variables in determining RT
# for stimulated trials across patients

lm.fit = lm(Zscore_RT~.-Latency_Preamble -Faster_RT -Trial, data=Cleaned_Stim_Trials) 
summary(lm.fit)

# check assumptions
library(ggfortify)
autoplot(lm.fit, label.size = 3)

# visualize in a 3-D manner
scatter3D(Cleaned_Stim_Trials$StimDuration, Cleaned_Stim_Trials$TimeofStim, Cleaned_Stim_Trials$Zscore_RT,
          phi=20, theta=20, xlab="StimDuration", ylab="TimeofStim",zlab="RT")
```

```r
# CLASSIFICATION 1

# Use glm to see what variables can predict if a given trial is going to be faster or slower than 'baseline'
library(ISLR)
library(ggplot2)

# Next let's quantify all pairwise correlations as a first glimpse at the relationships among the variables 
cxy = cor(Cleaned_Stim_Trials[,-9]) 
cxy

# Remove vairbales that deterimine Faster_RT as it is calculated, accuracy (all are one), and 
#PicCode - an aribitrary value assigned to each different image
glm.fit = glm(Faster_RT~.-Latency_Preamble -Zscore_RT -Patient -Trial, data=Cleaned_Stim_Trials, family=binomial) 
summary(glm.fit)

# Plot the model fit evaluation images
par(mfrow=c(2,2))
plot(glm.fit)

# Create training and test dataset to see how accurate at prediction
set.seed(23)   
train_ind  = sample(seq_len(nrow(Cleaned_Stim_Trials)),size = (.8*nrow(Cleaned_Stim_Trials)))
Data_train = Cleaned_Stim_Trials[train_ind,]
Data_test = Cleaned_Stim_Trials[-train_ind,]
FasterRT_test=Data_test$Faster_RT

# Run glm only on training data
glm.fit = glm(Faster_RT~.-Latency_Preamble -Zscore_RT -Patient, data=Cleaned_Stim_Trials, family=binomial,subset=train_ind) 
summary(glm.fit)

# Predict using glm.fit results in testing data
glm.probs=predict(glm.fit, Data_test, type="response")
glm.pred=rep("No",nrow(Data_test)) 
glm.pred[glm.probs>0.5]="Yes" 

confusion_df = data.frame(glm.pred, FasterRT_test) #create confusion df
colnames(confusion_df) = c('predicted', 'actual')
table(confusion_df)
mean(confusion_df$predicted == confusion_df$actual)

### PREDICTION ACCURACY = 68 %
```

```r
# CLASSIFICATION 2
# LDA and QDA
import_packages(c('MASS'))
library(MASS)


# remove familiarity due to fact that LDA doesnt like missing data + models above indicate not relevant for prediction
lda.fit = lda(Faster_RT~.-Latency_Preamble - Familiarity -Zscore_RT -Patient, data=Cleaned_Stim_Trials,subset=train_ind) 
lda.fit
plot(lda.fit)

lda.pred = predict(lda.fit, Data_test)
lda.class = lda.pred$class
table(lda.class, FasterRT_test)
lda.pred_accuracy = (54+16)/length(FasterRT_test)
lda.pred_accuracy

#Visualization of LDA for StimDuration
StimD_posterior_df = data.frame(Data_test$StimDuration, lda.pred$posterior[,1])
colnames(StimD_posterior_df) = c('Stim_Duration', 'Slower_posterior_prob')
ggplot(data=StimD_posterior_df, aes(Stim_Duration, Slower_posterior_prob)) + geom_point()

## 71% prediction accuracy with LDA


#QDA
qda.fit = qda(Faster_RT~.-Latency_Preamble - Familiarity -Zscore_RT -Patient, data=Cleaned_Stim_Trials,subset=train_ind) 
qda.fit

# Same basic predictions as LDA, just assuming that each
# distribution has its own variance. But we work with 
# coefficients the same way.
qda.class=predict(qda.fit, Data_test)$class
table(qda.class, FasterRT_test)
mean(qda.class==FasterRT_test)

#Visualization of QDA for StimDuration
StimD_posterior_df = data.frame(Data_test$StimDuration, qda.class)
colnames(StimD_posterior_df) = c('Stim_Duration', 'Faster_posterior_prob')
ggplot(data=StimD_posterior_df, aes(Stim_Duration, Faster_posterior_prob)) + geom_point()

## 76% prediction accuracy with QDA for StimDuration
```

```r
#For Permutaion resting - respampling without replacement
#-----------------------------------------------------------------------------------------------
#-----------------------------------------------------------------------------------------------
#-----------------------------------------------------------------------------------------------
  
   #REGRESSION PERMUTATION
library(MASS)

# Step 1: Calculate the real effect for StimDuration and TimeofStim in the mixed model

beta_real_StimDuration = lm(Cleaned_Stim_Trials$Zscore_RT~Cleaned_Stim_Trials$StimDuration + Cleaned_Stim_Trials$TimeofStim + Cleaned_Stim_Trials$Repetition)$coefficients[2]
beta_real_TimeofStim = lm(Cleaned_Stim_Trials$Zscore_RT~Cleaned_Stim_Trials$StimDuration + Cleaned_Stim_Trials$TimeofStim + Cleaned_Stim_Trials$Repetition)$coefficients[3]
#Step2: Run m=1000 permutations using the sample function to shuffle the variable of interest
beta_perm_Duration = matrix(NaN, nrow = 5000)
beta_perm_TimeofStim = matrix(NaN, nrow = 5000)
    #Shuffle just one variable at a time and create dataframe for each within forloop
for (i in 1:5000){
    beta_perm_Duration[i] = lm(Cleaned_Stim_Trials$Zscore_RT ~ sample(Cleaned_Stim_Trials$StimDuration, replace = FALSE) + Cleaned_Stim_Trials$TimeofStim + Cleaned_Stim_Trials$Repetition)$coefficients[2]
    beta_perm_TimeofStim[i] = lm(Cleaned_Stim_Trials$Zscore_RT ~ Cleaned_Stim_Trials$StimDuration + sample(Cleaned_Stim_Trials$TimeofStim, replace = FALSE) + Cleaned_Stim_Trials$Repetition)$coefficients[3]
}
    # See the real effects
beta_real_StimDuration
beta_real_StimDuration
    # See the max perm effects as ratio relative to real effects
max(beta_perm_Duration)/beta_real_StimDuration
max(beta_perm_TimeofStim)/beta_real_TimeofStim

#Step3: calculate the p value
p_duration = sum(beta_perm_Duration > beta_real_StimDuration)/5000
p_duration
p_timeofstim= sum(beta_perm_TimeofStim > beta_real_TimeofStim)/5000
p_timeofstim

## p_duration <.0002
## p_timeofstim = .005



#The bootstrap: resampling with replacement 
  #-----------------------------------------------------------------------------------------------
  #-----------------------------------------------------------------------------------------------
  #-----------------------------------------------------------------------------------------------
  
# Calculate the original effect

beta_real_StimDuration = lm(Cleaned_Stim_Trials$Zscore_RT~Cleaned_Stim_Trials$StimDuration + Cleaned_Stim_Trials$TimeofStim + Cleaned_Stim_Trials$Repetition)$coefficients[2]
beta_real_TimeofStim = lm(Cleaned_Stim_Trials$Zscore_RT~Cleaned_Stim_Trials$StimDuration + Cleaned_Stim_Trials$TimeofStim + Cleaned_Stim_Trials$Repetition)$coefficients[3]# Determine the sample size (will be important in a minute)
n = dim(Cleaned_Stim_Trials)[1]
#Creat a bootstrap of 1000 iteration of the regression model
boot_beta_Duration = matrix(NaN, nrow = 5000)
boot_beta_TimeofStim = matrix(NaN, nrow = 5000)
for (i in 1:5000){
  indx = sample(1:n, replace=TRUE)
  boot_beta_Duration[i] = lm(Cleaned_Stim_Trials$Zscore_RT[indx]~Cleaned_Stim_Trials$StimDuration[indx]+ Cleaned_Stim_Trials$TimeofStim[indx] + Cleaned_Stim_Trials$Repetition[indx])$coefficients[2]
  boot_beta_TimeofStim[i] = lm(Cleaned_Stim_Trials$Zscore_RT[indx]~Cleaned_Stim_Trials$StimDuration[indx]+ Cleaned_Stim_Trials$TimeofStim[indx] + Cleaned_Stim_Trials$Repetition[indx])$coefficients[3]
}

# Let's look at the bootstrap

hist(boot_beta_Duration)
hist(boot_beta_TimeofStim)

# We can compare the mean of the bootstrap with the original effet

beta_real_StimDuration # OG regression model
mean(boot_beta_Duration) # Mean of the bootstrap

beta_real_TimeofStim # OG regression model
mean(boot_beta_TimeofStim) # Mean of the bootstrap

# Finally we calculate the confidence intervals

     #for duration
s = sd(boot_beta_Duration)
m  = mean(boot_beta_Duration)
error_duration = qnorm(1-.025) * s/sqrt(n)
upper_bound_duration = m + error_duration
upper_bound_duration
lower_bound_duration = m - error_duration
lower_bound_duration
    #for time of timeofstim
s = sd(boot_beta_TimeofStim)
m  = mean(boot_beta_TimeofStim)
error_TimeofStim = qnorm(1-.025) * s/sqrt(n)
upper_bound_TimeofStim = m + error_TimeofStim
upper_bound_TimeofStim
lower_bound_TimeofStim = m - error_TimeofStim
lower_bound_TimeofStim

# 95% confidence interval of bootstrapping for stimduration 
    # (in multiple linear regression model) effect 0.00014563089766612 - 0.000141733188227489
# 95% confidence interval of bootstrapping for timeofstim 
      #in multiple linear regression model)  effect 5.70730667316688e-05 - 5.2245896502806e-05
```
