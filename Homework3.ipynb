```r
hist(floor(runif(10,1,7)), breaks=seq(0, 6, 1))
abline(h=10/6)

hist(floor(runif(1000,1,7)), breaks=seq(0, 6, 1))
abline(h=1000/6)

hist(floor(runif(10000,1,7)), breaks=seq(0, 6, 1))
abline(h=10000/6)
```

```r
hist(floor(runif(10000,1,7)), breaks=seq(0, 6, 1))

a<-floor(runif(10000,1,7))
b<-floor(runif(10000,1,7))
c<-a+b
hist((c), breaks=seq(0, 12, 1))
qqnorm(c, pch = 1, frame = FALSE)
qqline(c, col = "steelblue", lwd = 2)

die1<-floor(runif(10000,1,7))
die2<-floor(runif(10000,1,7))
die3<-floor(runif(10000,1,7))
c<-die1+die2+die3
hist((c), breaks=seq(0, 18, 1))
qqnorm(c, pch = 1, frame = FALSE)
qqline(c, col = "steelblue", lwd = 2)

die1<-floor(runif(10000,1,7))
die2<-floor(runif(10000,1,7))
die3<-floor(runif(10000,1,7))
die4<-floor(runif(10000,1,7))
die5<-floor(runif(10000,1,7))
die6<-floor(runif(10000,1,7))
c<-die1+die2+die3+die4+die5+die6
hist((c), breaks=seq(0, 36, 1))
qqnorm(c, pch = 1, frame = FALSE)
qqline(c, col = "steelblue", lwd = 2)



#The fourth setup - with the 6 die, is the most normal normal distrubution, as is made evident by the qqplots 
#which shows that its the plot with the hihghest intersection with the reference line on the normal Q-Qplot - thereby 
#prividing  evidence that it has the most normal normal distribution
```
